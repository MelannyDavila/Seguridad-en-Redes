### Búsqueda en Shodan
El método search retorna un **diccionario** con los resultados. Usamos la clave **total** para obtener el número de resultados. 

```python
# Usamos un try-catch para manejar los errores que sucedan en el bloque
try:
    # Realizar una búsqueda 
    results = api.search('phpmyadmin')
    # Mostrar los resultados 
    print('Resultados encontrados: {}'.format(results['total']))
    for result in results['matches']:
            print('IP: {}'.format(result['ip_str']))
            print('Localización:', result['location']['country_name'])
            print('Producto:', result.get('product', 'N/A'))
            print('Puerto:', result['port'])
            print('Organización:', result['org'])
            if 'http' in result and 'components' in result['http']:
                print('Componentes:', result['http']['components'])
            print('-----------------')
except shodan.APIError as e:
    print('Error: {}'.format(e))
```
