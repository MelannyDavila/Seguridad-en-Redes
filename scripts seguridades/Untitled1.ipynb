```python
# Usamos un try-catch para manejar los errores que sucedan en el bloque
try:
    # Realizar una búsqueda 
    results = api.search('190.15.136.171')
    # Mostrar los resultados 
    print('Resultados encontrados: {}'.format(results['total']))
    for result in results['matches']:
            print('Puertos:', result['port'])
            print('CVEs de las vulnerabilidades:')
            for cve in  result['vulns']:
                print(cve)
            print('-----------------')
except shodan.APIError as e:
    print('Error: {}'.format(e))
```
